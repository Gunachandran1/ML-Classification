```python
dataset=pd.get_dummies(dataset,drop_first=True,dtype=int)
```

```python
dataset.columns
```

```python
dataset.drop('age',axis=1,inplace=True)
```

```python
dataset.shape
```

```python
dataset.columns
```

```python
features=dataset[['bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv', 'wc',
       'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
target=dataset['classification_yes']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(features,target,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X_train=ss.fit_transform(X_train)
X_test=ss.transform(X_test)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
params={
    'C': [1, 10, 100],
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'], 
    'degree': [2, 3], 
    'gamma': ['scale', 'auto']
        }
grid=GridSearchCV(SVC(probability=True),params,refit=True,verbose=3,n_jobs=-1,scoring='f1_weighted')
```

```python
grid.fit(X_train,y_train)
```

```python
y_pred=grid.predict(X_test)
```

```python
grid.best_params_
```

```python
cl=grid.cv_results_
```

```python
pd.DataFrame.from_dict(cl)
```

```python
grid.best_params_
```

```python
from sklearn.metrics import confusion_matrix,classification_report
cm=confusion_matrix(y_test,y_pred)
```

```python
print(cm)
```

```python
cf=classification_report(y_test,y_pred)
```

```python
print(cf)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,grid.predict_proba(X_test)[:,1])
```
